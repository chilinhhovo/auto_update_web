## Setup: Import what you'll need to scrape the page

We'll be using either Playwrightfor this, *not* requests.

```python
from playwright.async_api import async_playwright
playwright = await async_playwright().start()
browser = await playwright.chromium.launch(headless=False)
page = await browser.new_page()
await page.goto("https://cbmanhattan.cityofnewyork.us/cb2/minutes/")
html = await page.content()
from bs4 import BeautifulSoup
```

## Starting your search

Starting from [here](https://www.nyc.gov/site/manhattancb1/archives/monthly-full-board-meeting-minutes.page), search for all community board meeting minutes from Community Board 1 Manhattan 

# Finding all minutes

```python
#Finding all the div of the minutes
from bs4 import BeautifulSoup
soup_doc = BeautifulSoup(html, "html.parser")
table = soup_doc.find_all("div", class_="span4")
table
```

```python
#getting text and all the links from the div
Manhattan_CB1 = {}
for a_tag in table: 
    tags = a_tag.find_all("a")
    for tag in tags: 
        href = tag.get("href")
        name = tag.text.strip()
        Manhattan_CB1[name] = href
        print(href)
        print(name)
```

```python
for name, url in Manhattan_CB1.items():
    print(f"{name}: {url}")
```

```python
import pandas as pd
df = pd.DataFrame(Manhattan_CB1.items(), columns=['Text', 'URL'])
df
```
