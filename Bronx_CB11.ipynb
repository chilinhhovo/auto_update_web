I stole this code from the internet and fixed it up, link [here](https://www.geeksforgeeks.org/downloading-pdfs-with-python-using-requests-and-beautifulsoup/)

```python
import os
import requests
from bs4 import BeautifulSoup

# URL from which PDFs are to be downloaded
url = "https://www.nyc.gov/site/bronxcb12/minutes/minutes.page"

# Requests URL and get response object
response = requests.get(url)

# Parse text obtained
soup = BeautifulSoup(response.text, 'html.parser')

# Find all hyperlinks present on the webpage
links = soup.find_all('a')

# Create a folder to save the PDF files
folder_name = "Bronx_CB11"
os.makedirs(folder_name, exist_ok=True)  # Create the folder if it doesn't exist

i = 0

# From all links, check for PDF links and download files
for link in links:
    if ('.pdf' in link.get('href', [])):
        i += 1
        print(f"Downloading file {i}...")

        # Construct the full URL
        pdf_url = "https://www.nyc.gov/" + link.get('href')

        # Extract the file name from the URL
        file_name = os.path.basename(link.get('href'))

        # Construct the path to save the file in the folder
        file_path = os.path.join(folder_name, file_name)

        # Get response object for the link
        response = requests.get(pdf_url)

        # Write content to the PDF file
        with open(file_path, 'wb') as pdf:
            pdf.write(response.content)

        print(f"File {i} downloaded and saved as: {file_path}")

print("All PDF files downloaded into the folder:", folder_name)
```
